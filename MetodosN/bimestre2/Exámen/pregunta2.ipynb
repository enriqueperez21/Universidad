```python
import numpy as np
import matplotlib.pyplot as plt

def fun(x):
    return (x**5)-6*(x**4)+2*(x**3)+20*(x**2)-27*x+10

x = np.linspace(-50, 50, 400)
y = fun(x)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, y, color='red', label='Datos predecidos')
plt.ylim(-400,300)
ax.legend()
```

```python
from scipy import optimize

def derivate_fun(x):
    return 5*(x**4)-24*(x**3)+6*(x**2)+40*x-27
```

```python
## optimize.newton(equation, 5)
optimize.newton(fun, [-50,-1,0,0,100], fprime= derivate_fun)
```

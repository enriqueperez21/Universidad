```python
import numpy as np

import matplotlib.pyplot as plt

def equation(x:float)->float:
    return (x**3 - 3*x**2 + x - 1)

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values
y = equation(x)

# Plot the equation
plt.plot(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of (x^3 - 3x^2 + x - 1) = 0$')
ax = plt.gca()
ax.set_ylim([-10, 10])
ax.set_xlim([-0.5, 4])
plt.grid(True)
plt.show()
```
